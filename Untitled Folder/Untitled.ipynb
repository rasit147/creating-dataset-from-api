```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage=requests.get('https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav',headers=headers).text
```

```python
soup=BeautifulSoup(webpage,'lxml')
```

```python
soup.find_all('h1')[0].text
```

```python
for i in soup.find_all('h2'):
    print(i.text.strip())
```

```python
company=soup.find_all('div',class_='company-content-wrapper')
```

```python
len(company)
```

```python
name=[]
rating=[]
reviews=[]
ctype=[]
hq=[]
old=[]
employees=[]


for i in company:
    name.append(i.find('h2').text.strip())
    rating.append(i.find('p',class_='rating').text.strip())
    reviews.append(i.find('a',class_='review-count').text.strip())
    ctype.append(i.find_all('p',class_='infoEntity')[2].text.strip())
    hq.append(i.find_all('p',class_='infoEntity')[2].text.strip())
    old.append(i.find_all('p',class_='infoEntity')[2].text.strip())
    employees.append(i.find_all('p',class_='infoEntity')[2].text.strip())
    
d={'name':name,'rating':rating,'reviews':reviews,'type':ctype,'hq':hq,'old':old,'employe':employees}
df=pd.DataFrame(d)
```
